This assignment focuses on the annotations done for the [Wikipedia Talk]( https://meta.wikimedia.org/wiki/Research:Detox) corpus, consisting of three datasets. The dataset contains 160k labeled comments in English Wikipedia from an original corpus of 64M comments from discussions relating to user pages and articles dating from 2004-2015. Each of the comments were labeled by at least 10 annotators via Crowdflower for the level of toxicity, aggression, and personal attack of a comment in progressing or ending a discourse. 

All the data collected by wikimedia can be accessed under free licenses via [figshare](https://figshare.com/articles/Wikipedia_Talk_Labels_Toxicity/4563973) and per wikimedia [open access policy](https://meta.wikimedia.org/wiki/Open_access_policy).

This notebook will be focusing its exploratory analysis on the [**aggression**](https://docs.google.com/spreadsheets/d/1IlhmnoAdHNm906WrwYp2el9tlZFfnJkwrYXzbEMKGvM/edit#gid=0&range=E3) and [**toxicity**](https://docs.google.com/spreadsheets/d/1IlhmnoAdHNm906WrwYp2el9tlZFfnJkwrYXzbEMKGvM/edit#gid=0&range=E2) datasets:

* **Aggression** : 100k labeled comments from English Wikipedia by approximately 10 annotators via Crowdflower on how aggressive the comment was perceived to be.
* **Toxicity** : 160k labeled comments from English Wikipedia by approximately 10 annotators via Crowdflower on a spectrum of how toxic the comment is (perceived as likely to make people want to leave the discussion) to how healthy to conversation the contribution is.

Each of the datasets contains the following subsets:
* Annotated Comments: The comments put to revision from the corpus
* Annotations: The labels given to comment and the different workers who labeled them
* Demographics: Demographics of gender, age group, English as first language, and education for the workers

## Select and Download data

We will now setup the URLs for the aggression and toxicity files to download from https://figshare.com/projects/Wikipedia_Talk/16731.

We will create a function to take these URLs and file names for where to download each of the files into a *data* folder in the repository. 

**NOTE** : This might take a few minutes to complete the download process

```python
# Data download URLs
AGGRESSION_ANNOTATED_COMMENTS_URL = 'https://ndownloader.figshare.com/files/7038038'
AGGRESSION_ANNOTATIONS_URL = 'https://ndownloader.figshare.com/files/7394506' 
AGGRESSION_WORKER_DEMOGRAPHICS_URL = 'https://ndownloader.figshare.com/files/7640644'

TOXICITY_ANNOTATED_COMMENTS_URL = 'https://ndownloader.figshare.com/files/7394542' 
TOXICITY_ANNOTATIONS_URL = 'https://ndownloader.figshare.com/files/7394539' 
TOXICITY_DEMOGRAPHICS_URL = 'https://ndownloader.figshare.com/files/7640581'

# files to download
files = {'aggression_annotated_comments':AGGRESSION_ANNOTATED_COMMENTS_URL,
         'aggression_annotations':AGGRESSION_ANNOTATIONS_URL,
         'aggression_worker_demographics':AGGRESSION_WORKER_DEMOGRAPHICS_URL,
         'toxicity_annotated_comments':TOXICITY_ANNOTATED_COMMENTS_URL,
         'toxicity_annotations':TOXICITY_ANNOTATIONS_URL,
         'toxicity_worker_demographics':TOXICITY_DEMOGRAPHICS_URL}

def download_data(url, file_name):
    """
        Downloads the data from Figshare.com and places it into "data" folder
    """
    if not os.path.exists('data'):
        os.makedirs('data')
        
    if not os.path.exists(file_name):
        urlretrieve(url, file_name)

for file, url in files.items():
    download_data(url,'data/'+file+'.tsv')
```

## Data Cleanup and Preparing for Analysis

We will now use pandas to read in the data from the tsv files into separate dataframe structures. We'll print one of these to check the general output of the dataframe.

```python
# loading data to dataframes
aggression_comments = pd.read_csv('data/aggression_annotated_comments.tsv', sep = '\t')
aggression_annotations = pd.read_csv('data/aggression_annotations.tsv',  sep = '\t')
aggression_demographics = pd.read_csv('data/aggression_worker_demographics.tsv',  sep = '\t')
toxicity_comments = pd.read_csv('data/toxicity_annotated_comments.tsv', sep = '\t')
toxicity_annotations = pd.read_csv('data/toxicity_annotations.tsv',  sep = '\t')
toxicity_demographics = pd.read_csv('data/toxicity_worker_demographics.tsv', sep = '\t')

# checking one df 
aggression_comments.head()
```

One thing to notice about the *comments* datesets is the fact that the actual comments have unnecessary, that being "NEWLINE_TOKEN" and "TAB_TOKEN". For now we will take these out, in case we want to look into the actual comments at a later time.

```python
# Remove newline and tab tokens from comments
aggression_comments['comment'] = aggression_comments['comment'].apply(lambda x: x.replace("NEWLINE_TOKEN", " "))
aggression_comments['comment'] = aggression_comments['comment'].apply(lambda x: x.replace("TAB_TOKEN", " "))
toxicity_comments['comment'] = toxicity_comments['comment'].apply(lambda x: x.replace("NEWLINE_TOKEN", " "))
toxicity_comments['comment'] = toxicity_comments['comment'].apply(lambda x: x.replace("TAB_TOKEN", " "))
```

With that out of the way, let's now just create the dataframe we want for our analysis purposes. We will merge the 3 *aggression* and *toxicity* into 2 dataframes containing the "rev_id", "worker_id", labels, scores, actual comment text, and demographics.

We'll merge first the demographics to the comments dataframe, to then merge this combined dataframe with the annotations dataframe to create the final **agg_df** and **tox_df** we will use for our analysis.

```python
agg_ac_df = pd.merge(aggression_annotations, aggression_comments[['rev_id','comment','year']], how='left', on=['rev_id'])
agg_df = pd.merge(agg_ac_df, aggression_demographics, how='left', on=['worker_id'])

tox_ac_df = pd.merge(toxicity_annotations, toxicity_comments[['rev_id','comment','year']], how='left', on=['rev_id'])
tox_df = pd.merge(tox_ac_df, toxicity_demographics, how='left', on=['worker_id'])
```

Let's just check that the dataframes look like we wanted and that they are about 100K and 160K comments (length) for the aggression and toxicity respectively.

## Analysis

### Part 1: Analysis on demographics

#### Analyze the demographic information about the Crowdflower workers that is available in the dataset

1. How evenly distributed are the gender and age demographics compared to the number of aggression comments reviewed?
2. Same question, but for the Toxicity dataset?

We will start by creating the necessary dataframes for the gender and age groupings for the Toxicity and Aggression labels.

```python
# generating the gender dataframe for aggression and toxicity
agg_gender_num_com = agg_df.groupby(['gender']).agg({'rev_id':'count'}).reset_index()
agg_gender_num_com['% of total Comments'] = agg_gender_num_com.rev_id/agg_gender_num_com.rev_id.sum()
agg_gender_num_com['Label Type'] = 'Aggression'
agg_gender_num_wkr = agg_df.groupby(['worker_id','gender']).rev_id.agg('count').reset_index(
    ).groupby('gender').worker_id.count().reset_index()
agg_gender_num_wkr['% of total workers'] = agg_gender_num_wkr.worker_id/agg_gender_num_wkr.worker_id.sum()
agg_gender_num_wkr['Label Type'] = 'Aggression'
agg_viz_gen_df = pd.merge(agg_gender_num_wkr, agg_gender_num_com, how='outer', on=['Label Type','gender'])

tox_gender_num_com = tox_df.groupby(['gender']).agg({'rev_id':'count'}).reset_index()
tox_gender_num_com['% of total Comments'] = tox_gender_num_com.rev_id/tox_gender_num_com.rev_id.sum()
tox_gender_num_com['Label Type'] = 'Toxicity'
tox_gender_num_wkr = tox_df.groupby(['worker_id','gender']).rev_id.agg('count').reset_index(
    ).groupby('gender').worker_id.count().reset_index()
tox_gender_num_wkr['% of total workers'] = tox_gender_num_wkr.worker_id/tox_gender_num_wkr.worker_id.sum()
tox_gender_num_wkr['Label Type'] = 'Toxicity'
tox_viz_gen_df = pd.merge(tox_gender_num_wkr, tox_gender_num_com, how='outer', on=['Label Type','gender'])

gender_viz = pd.concat([agg_viz_gen_df,tox_viz_gen_df])
gender_viz.rename(columns={"worker_id": "Number of Workers", "rev_id": "Reviewed Comments"}, errors="raise", inplace=True)

# generating the age dataframe for aggression and toxicity
agg_age_num_com = agg_df.groupby(['age_group']).agg({'rev_id':'count'}).reset_index()
agg_age_num_com['% of total Comments'] = agg_age_num_com.rev_id/agg_age_num_com.rev_id.sum()
agg_age_num_com['Label Type'] = 'Aggression'
agg_age_num_wkr = agg_df.groupby(['worker_id','age_group']).rev_id.agg('count').reset_index(
    ).groupby('age_group').worker_id.count().reset_index()
agg_age_num_wkr['% of total workers'] = agg_age_num_wkr.worker_id/agg_age_num_wkr.worker_id.sum()
agg_age_num_wkr['Label Type'] = 'Aggression'
agg_viz_age_df = pd.merge(agg_age_num_wkr, agg_age_num_com, how='outer', on=['Label Type','age_group'])

tox_age_num_com = tox_df.groupby(['age_group']).agg({'rev_id':'count'}).reset_index()
tox_age_num_com['% of total Comments'] = tox_age_num_com.rev_id/tox_age_num_com.rev_id.sum()
tox_age_num_com['Label Type'] = 'Toxicity'
tox_age_num_wkr = tox_df.groupby(['worker_id','age_group']).rev_id.agg('count').reset_index(
    ).groupby('age_group').worker_id.count().reset_index()
tox_age_num_wkr['% of total workers'] = tox_age_num_wkr.worker_id/tox_age_num_wkr.worker_id.sum()
tox_age_num_wkr['Label Type'] = 'Toxicity'
tox_viz_age_df = pd.merge(tox_age_num_wkr, tox_age_num_com, how='outer', on=['Label Type','age_group'])

# Final dataframe for visualization
age_viz = pd.concat([agg_viz_age_df,tox_viz_age_df])
age_viz.rename(columns={"worker_id": "Number of Workers", "rev_id": "Reviewed Comments"}, errors="raise", inplace=True)
```

```python
sns.set_style("white")
sns.set_palette("colorblind")
gen_viz_wkr = sns.catplot(
    data=gender_viz, kind="bar",
    x="gender", y="% of total workers", hue="Label Type",
    alpha=.8, height=6
)
gen_viz_cmt = sns.catplot(
    data=gender_viz, kind="bar",
    x="gender", y="% of total Comments", hue="Label Type",
    alpha=.8, height=6
)

age_viz_wkr = sns.catplot(
    data=age_viz, kind="bar",
    x="age_group", y="% of total workers", hue="Label Type",
    alpha=.8, height=6
)
age_viz_cmt = sns.catplot(
    data=age_viz, kind="bar",
    x="age_group", y="% of total Comments", hue="Label Type",
    alpha=.8, height=6
)

# save figures
gen_viz_cmt.savefig('Gender_Distribution_labels_Percent_comments.png')
gen_viz_wkr.savefig('Gender_Distribution_labels_Percent_workers.png')
age_viz_cmt.savefig('Age_Distribution_labels_Percent_comments.png')
age_viz_wkr.savefig('Age_Distribution_labels_Percent_workers.png')

# display figures
gen_viz_wkr
gen_viz_cmt
age_viz_wkr
age_viz_cmt
```

**Observations for Demographic breakdown for Gender and Age Groups**
* There is an unequal gender distribution in for both Aggression and Toxicity datasets, with about 60% of labelers being Male. The distribution is also seen in the number of comments being labeled by Male and Female workers.
* For the Age grouping, we see that the datasets are primarily composed of workers in between the ages of 18 to 45, with workers youger than 18 or older than 60 have a very low showing.
* The skew distribution for age groups and high inbalance for gender labelers are quite a source of concern for the bias it brings to the dataset, especially as to how it will affect algorithms trained on this dataset and how it deals with these less represented set of people.

### Part 2 : Exploration of Labeling and demographics
#### Explore relationships between worker demographics and labeling behavior

1. Are workers with Bachelors, Master's and Doctorate degrees more likely to label an a comment as aggression or toxic, than those with a different education degree?
2. Do we see an even distribution of rating for aggressive or toxic labels across age groups?

Let's now generate a dataframe with the desired output for the percentage of labels for the aggression and toxicity of a comment by Labelers with "Higher Degrees" (Bachelors, Masters, Doctorate) compared to those who do not have these degrees.

We will also generate the dataframes required for the analysis of the distribution of comments for toxicity and aggression labeling by age group.

```python
# generating toxicity dataframe for education
edu_higher_tox = pd.DataFrame(tox_df.loc[tox_df.education.isin(['bachelors','masters','doctorate']), 'toxicity_score'].
             value_counts(normalize=True)).stack().reset_index()
edu_higher_tox.rename(columns={'level_0':'Level','level_1':'Label',0:'Score'}, inplace=True)
edu_higher_tox['Education Level'] = 'Higher Degree'

edu_lower_tox = pd.DataFrame(tox_df.loc[~tox_df.education.isin(['bachelors','masters','doctorate']), 'toxicity_score'].
             value_counts(normalize=True)).stack().reset_index()
edu_lower_tox.rename(columns={'level_0':'Level','level_1':'Label',0:'Score'}, inplace=True)
edu_lower_tox['Education Level'] = 'Lower Degree'

edu_tox_df = pd.concat([edu_higher_tox, edu_lower_tox])

# generating aggression dataframe for education
edu_higher_agg = pd.DataFrame(agg_df.loc[agg_df.education.isin(['bachelors','masters','doctorate']), 'aggression_score'].
             value_counts(normalize=True)).stack().reset_index()
edu_higher_agg.rename(columns={'level_0':'Level','level_1':'Label',0:'Score'}, inplace=True)
edu_higher_agg['Education Level'] = 'Higher Degree'

edu_lower_agg = pd.DataFrame(agg_df.loc[~agg_df.education.isin(['bachelors','masters','doctorate']), 'aggression_score'].
             value_counts(normalize=True)).stack().reset_index()
edu_lower_agg.rename(columns={'level_0':'Level','level_1':'Label',0:'Score'}, inplace=True)
edu_lower_agg['Education Level'] = 'Lower Degree'

edu_agg_df = pd.concat([edu_higher_agg, edu_lower_agg])

# final Education dataframe
edu_df = pd.concat([edu_tox_df, edu_agg_df])


# Age group dataframe for distribution of Toxicity and Aggression labels
age_agg_df = pd.DataFrame(agg_df.groupby(['age_group']).
                          aggression_score.value_counts()/agg_df.groupby(['age_group']).
                          aggression_score.count()).rename(columns={'aggression_score':'% of Total'}).reset_index()
age_tox_df = pd.DataFrame(tox_df.groupby(['age_group']).
                          toxicity_score.value_counts()/tox_df.groupby(['age_group']).
                          toxicity_score.count()).rename(columns={'toxicity_score':'% of Total'}).reset_index()
```

Now we can visualize the distribution of how the comments were labeled by degree type.

```python
edu_df = pd.concat([edu_tox_df, edu_agg_df])
edu_aggtox_viz = sns.catplot(
    data=edu_df, kind="bar",
    x="Level", y="Score", hue='Education Level', col='Label',
    alpha=.8
)
plt.subplots_adjust(top=0.9)
edu_aggtox_viz.set_axis_labels("", "Percent of Total Comments")
edu_aggtox_viz.fig.suptitle('Distribution of Toxicity and Aggression Comments by Educational Level') 
edu_aggtox_viz.savefig('Distribution_of_Toxicity_Aggression_by_educational_group.png')
```

**Observations on Education's effect on labels:**
* The difference in the distribution for both Toxicity and Aggression labeling between "Higher" and "Lower" ed degrees is very similar and has little difference between the degree level.
* Overall, labelers are less likely to give a "Toxic" label to a comment, as seen by the distribution shifting more to the positive toxicity score (non-toxic label).
* Overall, for labeling aggression on comments, labelers seemed to be more likely to have a neutral view of comments with slight comfort to give an "aggressive" lebel as seen by the negative values having higher distribution compared to positive levels.


Now let's visualize the distribution of how the comments were labeled by age group.

First by level of Toxicity labels:

```python
tox_age_viz = sns.catplot(
    data=age_tox_df, kind="bar",
    x="toxicity_score", y="% of Total", col='age_group',col_wrap=3,
    alpha=.8
)

plt.subplots_adjust(top=0.9)
tox_age_viz.fig.suptitle('Distribution of Toxicity Level of comments by Age Group') 
tox_age_viz.savefig('Distribution_of_Toxicity_Level_of_comments_by_Age_Group.png')
```

Now let's visualize the Aggression label distribution:

```python
plt.subplots_adjust(top=0.9)
agg_age_viz = sns.catplot(
    data=age_agg_df, kind="bar",
    x="aggression_score", y="% of Total", col='age_group',col_wrap=3,
    alpha=.8
)
plt.subplots_adjust(top=0.9)
agg_age_viz.fig.suptitle('Distribution of Aggression Level of comments by Age Group') 
agg_age_viz.savefig('Distribution_of_Aggression_Level_of_comments_by_Age_Group.png')
```

**Observations on Age and Lebeling:**
* For Toxicity labeling, we see that workers tend to have a more positive view on the possible toxicity of a comment, as observed by the distributions being right skewed.
* For the Aggression dataset, we see again that labelers are more likely to have a neutral view overall regardless of age group. However, from glancing at the distributions, we can see that the age group "under 18" shows the least likelihood to label a comment as an aggression, while the groups over 45 show a shift towards lebeling more comments as aggression than not.
* Overall, we see that there is a shift towards labeling, possibly catching, a comment as negative in both aggression and toxicity as there is an increase in the age of a labeler especially compared to younger age groups. The dataset itset however is mostly composed by labelers in the 18-45 age ranges, which gives some balance to the two extremes in age ranges, but could be missing out on many negative labeling if the older age groups are correct in their labels.
